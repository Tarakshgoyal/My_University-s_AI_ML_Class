```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
data=pd.read_excel('lab4.xlsx')
X=data['X'].values
y=data['y'].values
mean_X = np.mean(X)
mean_y = np.mean(y)
num = np.sum((X - mean_X) * (y - mean_y))
denom = np.sum((X - mean_X)**2)
b1 = num / denom 
b0 = mean_y - b1 * mean_X 
y_pred = b0 + b1 * X
x_pred1=1100
x_pred2=2575
y_pred1=b0+b1*x_pred1
y_pred2=b0+b1*x_pred2
print("Predicted value for x=1100 is", y_pred1)
print("Predicted value for x=2575 is",y_pred2)
plt.scatter(X, y, color='blue')
plt.plot(X, y_pred, color='red')
plt.xlabel('X')
plt.ylabel('y')
plt.show()
```
