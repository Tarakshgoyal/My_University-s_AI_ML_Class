```python
import numpy as np
from sklearn.decomposition import PCA
```

```python
X = np.array([[2, 3],
              [3, 4],
              [4, 5],
              [5, 6]], dtype=float)
pca = PCA(n_components=2)
Z = pca.fit_transform(X) 
axes_sklearn = pca.components_   
ev = pca.explained_variance_
evr = pca.explained_variance_ratio_
```

```python
print("axis 1", np.round(axes_sklearn[0], 8))
print("axis 2", np.round(axes_sklearn[1], 8))
print("Projections on PC1 :", np.round(Z[:, 0], 8))
```

```python
mu = X.mean(axis=0)
Xc = X - mu
C = np.cov(Xc, rowvar=False, bias=False)
w, V = np.linalg.eigh(C)    
idx = np.argsort(w)[::-1]
w = w[idx]
V = V[:, idx]             
u1 = V[:, 0]            
u2 = V[:, 1]           
proj_pc1 = np.sum(Xc * u1, axis=1)
```

```python

print("Mean:", mu)
print("Covariance matrix:\n", np.round(C, 8))
print("Eigenvalues :", np.round(w, 8))
print("axis 1", np.round(u1, 8))
print("axis 2", np.round(u2, 8))
print("Projections on PC1", np.round(proj_pc1, 8))
```
