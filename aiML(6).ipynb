```python
def one_hot_encode(data):
    encoded_df = pd.DataFrame(index=data.index)
    categorical_cols = data.select_dtypes(include=['object']).columns
    if len(categorical_cols) == 0:
        return data
    for col in categorical_cols:
        unique_values = data[col].unique()
        for value in unique_values:
            col_name = f"{col}_{value}"
            encoded_df[col_name] = (data[col] == value).astype(int)
    numerical_cols = data.select_dtypes(exclude=['object']).columns
    final_df = pd.concat([data[numerical_cols], encoded_df], axis=1)
    return final_df
```

```python
f_df = one_hot_encode(data)
```

```python
data = {
    'x': [1,2,3,2,5,8,11],
    'y': [40,45,50,48,55,70,80]
}
df = pd.DataFrame(data)
correlation = df['x'].corr(df['y'])
covariance = df['x'].cov(df['y'])
print(f"Correlation between x and y: {correlation:.4f}")
print(f"Covariance between x and y: {covariance:.4f}")
correlation_matrix = df.corr()
print("\nCorrelation Matrix:")
print(correlation_matrix)
covariance_matrix = df.cov()
print("\nCovariance Matrix:")
print(covariance_matrix)
```
