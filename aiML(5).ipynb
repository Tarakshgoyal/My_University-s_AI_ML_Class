```python
def min_max_normalize(values, new_min=0, new_max=1):
    old_min = min(values)
    old_max = max(values)
    normalized = [
        ((v - old_min) / (old_max - old_min)) * (new_max - new_min) + new_min
        for v in values
    ]
    return normalized

data = [5, 10, 15, 20, 2, 22]
normalized_data = min_max_normalize(data, 0, 1)

print("Original:", data)
print("Normalized:", normalized_data)
```
